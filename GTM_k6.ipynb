```python
import pandas as pd
import numpy as np
from ugtm import eGTM, eGTC
import altair as alt
```

```python
# 파일 경로 설정
path = "/Users/woosu/Desktop/project/"



```

```python
# CSV 파일 읽기
document_topic_assignments = pd.read_csv(path + "document_topic_assignments_k6.csv")
topic_distributions = pd.read_csv(path + "topic_distributions_k6.csv")

# 문서별 토픽 할당 추출
topic_assignments = document_topic_assignments['topic'].values

# LDA 토픽 분포 추출
topic_distributions = topic_distributions.drop(columns=['doc_id']).values

# GTM 모델 학습 및 변환 
# 주어진 항만 특허 데이터의 주제 Scope를 고려하여 적합한 차원 수 11로 설정
gtm_model = eGTM(k=11, model="responsibilities").fit(topic_distributions)
gtm_means = eGTM(k=11, model="means").fit_transform(topic_distributions)
gtm_modes = eGTM(k=11, model="modes").fit_transform(topic_distributions)
```

```python

# GTM 결과 시각화
dgtm_modes = pd.DataFrame(gtm_modes, columns=["x1", "x2"])
dgtm_modes["topic"] = topic_assignments

gtm_modes_chart = alt.Chart(dgtm_modes).mark_circle().encode(
    x='x1',
    y='x2',
    color=alt.Color('topic:N', scale=alt.Scale(scheme='viridis')),
    size=alt.value(50),
    tooltip=['x1', 'x2', 'topic:N']
).properties(title="GTM (modes)", width=400, height=400)

dgtm_means = pd.DataFrame(gtm_means, columns=["x1", "x2"])
dgtm_means["topic"] = topic_assignments

gtm_means_chart = alt.Chart(dgtm_means).mark_circle().encode(
    x='x1',
    y='x2',
    color=alt.Color('topic:N', scale=alt.Scale(scheme='viridis')),
    size=alt.value(50),
    tooltip=['x1', 'x2', 'topic:N']
).properties(title="GTM (means)", width=400, height=400)
```

```python
# 두 시각화 결합
gtm_chart = gtm_means_chart | gtm_modes_chart
```

```python

# Construct GTM
gtm_all = eGTM(k=11, model="responsibilities").fit(topic_distributions)

gtm_modes_all = eGTM(k=11, model="modes").fit_transform(topic_distributions)

dgtm_modes_all = pd.DataFrame(gtm_modes_all, columns=["x1", "x2"])
# dgtm_modes["label"] = y

gtm_modes_all = alt.Chart(dgtm_modes_all).mark_circle().encode(
    x='x1',
    y='x2',
#    color=alt.Color('label:N',
#                    scale=alt.Scale(scheme='viridis')),
    size=alt.value(3000),
    tooltip=['x1','x2','label:N']
).properties(title = "GTM (modes)", width = 500, height = 500)



```
